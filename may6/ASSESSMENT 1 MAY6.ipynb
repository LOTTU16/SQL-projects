```python
from pyspark.sql.functions import *

# 11. Convert registration_date to Date
patients_df = patients_df.withColumn(
    "registration_date",
    to_date("registration_date")
)

# 12. Convert visit_date to Date
visits_df = visits_df.withColumn(
    "visit_date",
    to_date("visit_date")
)

# 13. Add test_cost column
visits_df = visits_df.withColumn(
    "test_count",
    col("tests_count") * 500
)

# 14. Add estimated_total_bill column
visits_df = visits_df.withColumn(
    "estimated_total_bill",
    col("test_count") + 1000
)

# 15. Create age_category
patients_df = patients_df.withColumn(
    "age_category",
    when(col("age") < 30, "Young")
    .when(col("age") < 50, "Middle")
    .otherwise("Senior")
)

# 16. Create membership_priority
patients_df = patients_df.withColumn(
    "membership_priority",
    when(col("membership") == "Gold", 1)
    .when(col("membership") == "Silver", 2)
    .otherwise(3)
)

# 17. Create visit_priority
visits_df = visits_df.withColumn(
    "visit_priority",
    when(col("visit_status") == "Completed", 1)
    .when(col("visit_status") == "Pending", 2)
    .otherwise(3)
)

# 18. Create high_bill_flag
visits_df = visits_df.withColumn(
    "high_bill_flag",
    when(col("estimated_total_bill") > 3000, "Yes")
    .otherwise("No")
)

# 19. Rename patient_name to full_name
patients_df = patients_df.withColumnRenamed(
    "patient_name",
    "full_name"
)

# 20. (Optional cleanup - if any temp columns)
# Example: drop if not needed
# visits_df = visits_df.drop("test_cost")

# Final check
display(patients_df)
display(visits_df)
```

```python
from pyspark.sql.functions import *

# 21. Join patients with visits
pv_df = patients_df.join(visits_df, "patient_id")

# 22. Join visits with doctors
pvd_df = pv_df.join(doctors_df, "doctor_id")

# 23. Join visits with payments
final_df = pvd_df.join(payments_df, "visit_id")

# 24. Final joined DataFrame
display(final_df)

# 25. Required columns
final_df.select(
    "full_name",
    "city",
    "doctor_name",
    "specialization",
    "visit_status",
    "bill_amount"
).show()

# 26. Patient city != Doctor city
final_df.filter(
    col("city") != col("doctor_city")
).show()

# 27. Completed visits + Paid payments
final_df.filter(
    (col("visit_status") == "Completed") &
    (col("payment_status") == "Paid")
).show()

# 28. Pending visits + Pending payments
final_df.filter(
    (col("visit_status") == "Pending") &
    (col("payment_status") == "Pending")
).show()

# 29. Cancelled visits + Cancelled payments
final_df.filter(
    (col("visit_status") == "Cancelled") &
    (col("payment_status") == "Cancelled")
).show()

# 30. Patients who visited more than once
final_df.groupBy("patient_id") \
    .count() \
    .filter(col("count") > 1) \
    .show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

# 51. Rank patients by total bill within city
w1 = Window.partitionBy("city").orderBy(desc("bill_amount"))

df_rank_patient = final_df.withColumn(
    "patient_rank",
    rank().over(w1)
)
display(df_rank_patient)

# 52. Rank doctors by total revenue within specialization
w2 = Window.partitionBy("specialization").orderBy(desc("bill_amount"))

df_rank_doctor = final_df.withColumn(
    "doctor_rank",
    rank().over(w2)
)
display(df_rank_doctor)

# 53. ROW_NUMBER → top billing patient per city
w3 = Window.partitionBy("city").orderBy(desc("bill_amount"))

df_top_patient = final_df.withColumn(
    "row_num",
    row_number().over(w3)
).filter("row_num = 1")

display(df_top_patient)

# 54. DENSE_RANK doctors by consultation fee within specialization
w4 = Window.partitionBy("specialization").orderBy(desc("consultation_fee"))

df_dense = final_df.withColumn(
    "dense_rank",
    dense_rank().over(w4)
)
display(df_dense)

# 55. Top 2 doctors by revenue
w5 = Window.orderBy(desc("bill_amount"))

df_top2 = final_df.withColumn(
    "rank",
    rank().over(w5)
).filter("rank <= 2")

display(df_top2)

# 56. Running total revenue by visit date
w6 = Window.orderBy("visit_date")

df_running_date = final_df.withColumn(
    "running_total",
    sum("bill_amount").over(w6)
)
display(df_running_date)

# 57. Running total revenue by doctor
w7 = Window.partitionBy("doctor_name").orderBy("visit_date")

df_running_doc = final_df.withColumn(
    "doctor_running_total",
    sum("bill_amount").over(w7)
)
display(df_running_doc)

# 58. Rank cities by total revenue
city_df = final_df.groupBy("city") \
    .agg(sum("bill_amount").alias("total_revenue"))

w8 = Window.orderBy(desc("total_revenue"))

city_rank = city_df.withColumn(
    "rank",
    rank().over(w8)
)
display(city_rank)

# 59. Rank specializations by total revenue
spec_df = final_df.groupBy("specialization") \
    .agg(sum("bill_amount").alias("total_revenue"))

w9 = Window.orderBy(desc("total_revenue"))

spec_rank = spec_df.withColumn(
    "rank",
    rank().over(w9)
)
display(spec_rank)

# 60. Highest bill per payment mode
w10 = Window.partitionBy("payment_mode").orderBy(desc("bill_amount"))

df_highest = final_df.withColumn(
    "row_num",
    row_number().over(w10)
).filter("row_num = 1")

display(df_highest)
```

```python
from pyspark.sql.functions import *

daily_visits_data = [
(26,1002,201,"2024-03-14","Completed",2),
(27,1005,210,"2024-03-14","Completed",1),
(11,1011,205,"2024-03-06","Completed",3),
(20,1020,206,"2024-03-10","Completed",2),
(28,1018,203,"2024-03-15","Pending",2)
]

daily_visits_columns = [
"visit_id","patient_id","doctor_id","visit_date","visit_status","tests_count"
]

daily_visits_df = spark.createDataFrame(daily_visits_data, daily_visits_columns)

# Convert date
daily_visits_df = daily_visits_df.withColumn("visit_date", to_date("visit_date"))

display(daily_visits_df)
```

```python
daily_visits_df.createOrReplaceTempView("daily_updates")
```

```python
%sql
MERGE INTO hospital_db.visits_target AS target
USING daily_updates AS source
ON target.visit_id = source.visit_id

WHEN MATCHED THEN
UPDATE SET
target.patient_id = source.patient_id,
target.doctor_id = source.doctor_id,
target.visit_date = source.visit_date,
target.visit_status = source.visit_status,
target.tests_count = source.tests_count

WHEN NOT MATCHED THEN
INSERT (
visit_id, patient_id, doctor_id, visit_date, visit_status, tests_count
)
VALUES (
source.visit_id, source.patient_id, source.doctor_id,
source.visit_date, source.visit_status, source.tests_count
)
```
