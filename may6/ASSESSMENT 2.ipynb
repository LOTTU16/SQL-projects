```python
from pyspark.sql.functions import *

# 11. Convert order_date to date
orders_df = orders_df.withColumn(
    "order_date",
    to_date(col("order_date"), "yyyy-MM-dd")
)

# 12. Add total_order_value (join with products)
orders_with_price = orders_df.join(products_df, "product_id")

orders_with_price = orders_with_price.withColumn(
    "total_order_value",
    col("quantity") * col("price")
)

# 13. Create stock_status
products_df = products_df.withColumn(
    "stock_status",
    when(col("stock_quantity") < 20, "Low")
    .when(col("stock_quantity") < 50, "Medium")
    .otherwise("High")
)

# 14. Create order_priority
orders_df = orders_df.withColumn(
    "order_priority",
    when(col("quantity") > 50, "High")
    .when(col("quantity") > 20, "Medium")
    .otherwise("Low")
)

# 15. Create expensive_product_flag
products_df = products_df.withColumn(
    "expensive_product_flag",
    when(col("price") > 20000, "Yes").otherwise("No")
)

# 16. Convert category to uppercase
products_df = products_df.withColumn(
    "category",
    upper(col("category"))
)

# 17. Rename warehouse_city to inventory_city
products_df = products_df.withColumnRenamed(
    "warehouse_city",
    "inventory_city"
)

# 18. Create inventory_value
products_df = products_df.withColumn(
    "inventory_value",
    col("price") * col("stock_quantity")
)

# 19. Drop temporary column (price from joined df if needed)
orders_with_price = orders_with_price.drop("price")

# 20. Create low_stock_flag
products_df = products_df.withColumn(
    "low_stock_flag",
    when(col("stock_quantity") < 20, "Yes").otherwise("No")
)
```

```python
from pyspark.sql.functions import *

# 21. Join products with orders
prod_orders_df = orders_df.join(products_df, "product_id", "inner")

# 22. Join orders with suppliers
orders_suppliers_df = orders_df.join(suppliers_df, "supplier_id", "inner")

# 23. Join orders with payments
orders_payments_df = orders_df.join(payments_df, "order_id", "inner")

# 24. Final joined DataFrame (ALL tables)
final_df = orders_df \
    .join(products_df, "product_id") \
    .join(suppliers_df, "supplier_id") \
    .join(payments_df, "order_id")

# 25. Display required columns
final_df.select(
    "product_name",
    "supplier_name",
    "quantity",
    "bill_amount"
).show()

# 26. Suppliers serving different cities
suppliers_df.select("supplier_name","supplier_city").distinct().show()

# 27. Delivered orders with Paid payments
final_df.filter(
    (col("order_status") == "Delivered") &
    (col("payment_status") == "Paid")
).show()

# 28. Pending orders with Pending payments
final_df.filter(
    (col("order_status") == "Pending") &
    (col("payment_status") == "Pending")
).show()

# 29. Cancelled orders with Cancelled payments
final_df.filter(
    (col("order_status") == "Cancelled") &
    (col("payment_status") == "Cancelled")
).show()

# 30. Products ordered multiple times
final_df.groupBy("product_name") \
    .count() \
    .filter(col("count") > 1) \
    .show()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

# 51. Rank products by revenue within category
win_cat = Window.partitionBy("category").orderBy(col("bill_amount").desc())

final_df = final_df.withColumn(
    "product_rank_in_category",
    rank().over(win_cat)
)

# 52. Rank suppliers by revenue within city
win_city = Window.partitionBy("supplier_city").orderBy(col("bill_amount").desc())

final_df = final_df.withColumn(
    "supplier_rank_in_city",
    rank().over(win_city)
)

# 53. ROW_NUMBER → top product per category
win_row = Window.partitionBy("category").orderBy(col("bill_amount").desc())

top_product_per_category = final_df.withColumn(
    "row_num",
    row_number().over(win_row)
).filter(col("row_num") == 1)

top_product_per_category.show()

# 54. DENSE_RANK suppliers by bill amount (within city)
final_df = final_df.withColumn(
    "supplier_dense_rank",
    dense_rank().over(win_city)
)

# 55. Top 2 suppliers by revenue (FIXED with partition)
win_sup_rank = Window.partitionBy("supplier_city").orderBy(col("bill_amount").desc())

top2_suppliers = final_df.withColumn(
    "rank_sup",
    rank().over(win_sup_rank)
).filter(col("rank_sup") <= 2)

top2_suppliers.show()

# 56. Running total revenue by order date (FIXED with partition)
win_date = Window.partitionBy("supplier_name") \
    .orderBy("order_date") \
    .rowsBetween(Window.unboundedPreceding, 0)

final_df = final_df.withColumn(
    "running_total_revenue",
    sum("bill_amount").over(win_date)
)

# 57. Running total revenue by supplier (already correct)
win_sup = Window.partitionBy("supplier_name") \
    .orderBy("order_date") \
    .rowsBetween(Window.unboundedPreceding, 0)

final_df = final_df.withColumn(
    "running_supplier_revenue",
    sum("bill_amount").over(win_sup)
)

# 58. Rank cities by revenue
city_revenue = final_df.groupBy("supplier_city") \
    .agg(sum("bill_amount").alias("total_revenue"))

city_revenue = city_revenue.withColumn(
    "city_rank",
    rank().over(Window.orderBy(col("total_revenue").desc()))
)

city_revenue.show()

# 59. Rank categories by revenue
cat_revenue = final_df.groupBy("category") \
    .agg(sum("bill_amount").alias("total_revenue"))

cat_revenue = cat_revenue.withColumn(
    "category_rank",
    rank().over(Window.orderBy(col("total_revenue").desc()))
)

cat_revenue.show()

# 60. Highest bill order per payment mode
win_pay = Window.partitionBy("payment_mode").orderBy(col("bill_amount").desc())

top_bill_per_mode = final_df.withColumn(
    "row_num",
    row_number().over(win_pay)
).filter(col("row_num") == 1)

top_bill_per_mode.show()
```

```python
from pyspark.sql.functions import *

# 71. Create daily updates DataFrame
daily_orders_data = [
(321,114,204,"2024-04-11",3,"Delivered"),
(322,118,210,"2024-04-11",2,"Delivered"),
(304,114,208,"2024-04-02",5,"Delivered"),  # existing → update
(319,112,208,"2024-04-10",2,"Delivered"),  # existing → update
(323,120,210,"2024-04-12",1,"Pending")     # new → insert
]

daily_orders_columns = [
"order_id",
"product_id",
"supplier_id",
"order_date",
"quantity",
"order_status"
]

daily_orders_df = spark.createDataFrame(daily_orders_data, daily_orders_columns)

# 72. Create target Delta table
orders_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("orders_target")

# 73. Create temp view for updates
daily_orders_df.createOrReplaceTempView("daily_updates")

# 74–76. MERGE (UPDATE + INSERT)
spark.sql("""
MERGE INTO orders_target t
USING daily_updates s
ON t.order_id = s.order_id

WHEN MATCHED THEN
  UPDATE SET
    t.product_id = s.product_id,
    t.supplier_id = s.supplier_id,
    t.order_date = s.order_date,
    t.quantity = s.quantity,
    t.order_status = s.order_status

WHEN NOT MATCHED THEN
  INSERT (
    order_id,
    product_id,
    supplier_id,
    order_date,
    quantity,
    order_status
  )
  VALUES (
    s.order_id,
    s.product_id,
    s.supplier_id,
    s.order_date,
    s.quantity,
    s.order_status
  )
""")

# 77. Verify updated records
spark.sql("""
SELECT * FROM orders_target
WHERE order_id IN (304,319)
""").show()

# 78. Verify inserted records
spark.sql("""
SELECT * FROM orders_target
WHERE order_id IN (321,322,323)
""").show()

# 79. Check Delta history
spark.sql("DESCRIBE HISTORY orders_target").show()
```
