```python
# 1–10 Basic Exploration

# 1. Display DataFrame
display(df)

# 2. Print schema
df.printSchema()

# 3. Count records
print("Total Records:", df.count())

# 4. First 5 rows
df.show(5)

# 5. Last 5 rows
df.orderBy("emp_id", ascending=False).show(5)

# 6. Column names
print(df.columns)

# 7. Select name and salary
df.select("name", "salary").show()

# 8. Select name, department, experience
df.select("name", "department", "experience").show()

# 9. Distinct departments
df.select("department").distinct().show()

# 10. Distinct cities
df.select("city").distinct().show()
```

```python
# 11–20 Filtering

# 11 Salary > 90000
df.filter(df.salary > 90000).show()

# 12 Engineering
df.filter(df.department == "Engineering").show()

# 13 Marketing
df.filter(df.department == "Marketing").show()

# 14 Experience > 6
df.filter(df.experience > 6).show()

# 15 Salary < 70000
df.filter(df.salary < 70000).show()

# 16 London
df.filter(df.city == "London").show()

# 17 Finance & salary > 90000
df.filter((df.department == "Finance") & (df.salary > 90000)).show()

# 18 Engineering & experience > 5
df.filter((df.department == "Engineering") & (df.experience > 5)).show()

# 19 Salary between 85000 and 95000
df.filter(df.salary.between(85000, 95000)).show()

# 20 Experience < 4
df.filter(df.experience < 4).show()
```

```python
from pyspark.sql.functions import col, lit

# 31–40 Transformations

df2 = df.withColumn("annual_salary", col("salary") * 12) \
.withColumn("salary_in_thousands", col("salary") / 1000) \
.withColumn("experience_double", col("experience") * 2) \
.withColumn("company", lit("GlobalTech")) \
.withColumn("salary_after_bonus", col("salary") + 5000) \
.withColumn("salary_after_tax", col("salary") * 0.9) \
.withColumn("experience_next_year", col("experience") + 1) \
.withColumn("salary_per_year_of_experience", col("salary") / col("experience"))

# 39 Rename salary → monthly_salary
df2 = df2.withColumnRenamed("salary", "monthly_salary")

# 40 Drop experience_double
df2 = df2.drop("experience_double")

df2.show()
```

```python
from pyspark.sql.functions import sum, avg, max, min, count

# 41–50 Aggregations

# Total salary
df.agg(sum("salary")).show()

# Average salary
df.agg(avg("salary")).show()

# Max salary
df.agg(max("salary")).show()

# Min salary
df.agg(min("salary")).show()

# Count employees
df.agg(count("*")).show()

# Count per department
df.groupBy("department").count().show()

# Avg salary per department
df.groupBy("department").agg(avg("salary")).show()

# Max salary per department
df.groupBy("department").agg(max("salary")).show()

# Min salary per department
df.groupBy("department").agg(min("salary")).show()

# Total salary per department
df.groupBy("department").agg(sum("salary")).show()
```

```python
# 51–60 GroupBy

# Count per city
df.groupBy("city").count().show()

# Avg experience per department
df.groupBy("department").agg(avg("experience")).show()

# Max experience per department
df.groupBy("department").agg(max("experience")).show()

# Total salary per city
df.groupBy("city").agg(sum("salary")).show()

# Avg salary per city
df.groupBy("city").agg(avg("salary")).show()

# Highest salary per city
df.groupBy("city").agg(max("salary")).show()

# Lowest salary per city
df.groupBy("city").agg(min("salary")).show()

# Avg experience per city
df.groupBy("city").agg(avg("experience")).show()

# Count per department & city
df.groupBy("department", "city").count().show()

# Total salary per department & city
df.groupBy("department", "city").agg(sum("salary")).show()
```

```python
from pyspark.sql.functions import when

# 61 Remove duplicates
df_clean = df.dropDuplicates()

# 62 Check nulls
df_clean.select([count(when(col(c).isNull(), c)).alias(c) for c in df_clean.columns]).show()

# 63 Replace null salary with 0
df_clean = df_clean.fillna({"salary": 0})

# 64 Drop rows where department is null
df_clean = df_clean.dropna(subset=["department"])

# 65 Fill missing experience with avg
avg_exp = df_clean.select(avg("experience")).first()[0]
df_clean = df_clean.fillna({"experience": avg_exp})

# 66 Row count after cleaning
print("Rows after cleaning:", df_clean.count())

# 67 Replace null department with "Unknown"
df_clean = df_clean.fillna({"department": "Unknown"})

# 68 Remove salary < 60000
df_clean = df_clean.filter(col("salary") >= 60000)

# 69 Remove experience < 2
df_clean = df_clean.filter(col("experience") >= 2)

# 70 Show cleaned data
df_clean.show()
```
