```sql
-- 16
SELECT AVG(salary) AS avg_salary FROM employees;

-- 17
SELECT MAX(salary) AS max_salary FROM employees;

-- 18
SELECT MIN(salary) AS min_salary FROM employees;

-- 19
SELECT department, COUNT(*) AS emp_count
FROM employees
GROUP BY department;

-- 20
SELECT department, AVG(salary) AS avg_salary
FROM employees
GROUP BY department;
```

```sql
-- 21
SELECT department, SUM(salary) AS total_salary
FROM employees
GROUP BY department;

-- 22
SELECT department, AVG(experience) AS avg_experience
FROM employees
GROUP BY department;

-- 23
SELECT city, COUNT(*) AS total_employees
FROM employees
GROUP BY city;

-- 24
SELECT city, MAX(salary) AS max_salary
FROM employees
GROUP BY city;

-- 25
SELECT department, MIN(salary) AS min_salary
FROM employees
GROUP BY department;
```

```sql
-- 26
SELECT *
FROM employees
WHERE salary > (SELECT AVG(salary) FROM employees);

-- 27
SELECT *
FROM employees e
WHERE experience > (
    SELECT AVG(experience)
    FROM employees
    WHERE department = e.department
);

-- 28
SELECT * FROM employees WHERE salary < 70000;

-- 29
SELECT * FROM employees WHERE department = 'HR';

-- 30
SELECT *
FROM employees
WHERE department = 'Finance' AND salary > 90000;
```

```sql
-- 31
SELECT e.*, p.budget
FROM employees e
JOIN projects p
ON e.department = p.department;

-- 32
SELECT e.emp_name, e.department, p.project_name
FROM employees e
JOIN projects p
ON e.department = p.department;

-- 33
SELECT department, SUM(budget) AS total_budget
FROM projects
GROUP BY department;

-- 34
SELECT DISTINCT e.*
FROM employees e
JOIN projects p
ON e.department = p.department;

-- 35
SELECT p.project_name, COUNT(e.emp_id) AS emp_count
FROM projects p
JOIN employees e
ON p.department = e.department
GROUP BY p.project_name;
```

```sql
-- 36
SELECT *
FROM employees
WHERE department = 'Engineering'
ORDER BY salary DESC
LIMIT 3;

-- 37
SELECT department
FROM employees
GROUP BY department
ORDER BY AVG(salary) DESC
LIMIT 1;

-- 38
SELECT city
FROM employees
GROUP BY city
ORDER BY COUNT(*) DESC
LIMIT 1;

-- 39
SELECT *
FROM employees e
WHERE salary > (
    SELECT AVG(salary)
    FROM employees
    WHERE department = e.department
);

-- 40
SELECT *
FROM employees e
WHERE experience = (
    SELECT MAX(experience)
    FROM employees
    WHERE department = e.department
);
```

```sql
CREATE OR REPLACE TABLE employee_salary_analysis AS
SELECT 
    emp_name,
    city,
    department,
    salary,
    experience,

    CASE
        WHEN salary >= 95000 THEN 'High'
        WHEN salary >= 80000 THEN 'Medium'
        ELSE 'Low'
    END AS salary_category,

    AVG(salary) OVER (PARTITION BY department) AS department_avg_salary

FROM employees;
```
