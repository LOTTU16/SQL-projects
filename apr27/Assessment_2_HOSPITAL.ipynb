```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Hospital Project").getOrCreate()

patients = spark.read.csv("patients.csv", header=True, inferSchema=True)
doctors = spark.read.csv("doctors.csv", header=True, inferSchema=True)
appointments = spark.read.csv("appointments.csv", header=True, inferSchema=True)
bills = spark.read.csv("bills.csv", header=True, inferSchema=True)
profiles = spark.read.json("patient_profiles.json", multiLine=True)
```

```python
# 1. Show all patients
patients.show()

# 2. Show all doctors
doctors.show()

# 3. Show all appointments
appointments.show()

# 4. Show all bills
bills.show()

# 5. Print schema of patients
patients.printSchema()

# 6. Print schema of doctors
doctors.printSchema()

# 7. Print schema of appointments
appointments.printSchema()

# 8. Count total patients
print("Total Patients:", patients.count())

# 9. Count total doctors
print("Total Doctors:", doctors.count())

# 10. Show first 5 appointments
appointments.show(5)
```

```python
from pyspark.sql.functions import col

# 11. Select patient name, city, and age
patients.select("name", "city", "age").show()

# 12. Select doctor name, specialization, and consultation fee
doctors.select("doctor_name", "specialization", "consultation_fee").show()

# 13. Rename name in patients to patient_name
patients.withColumnRenamed("name", "patient_name").show()

# 14. Rename doctor_name to consultant_name
doctors.withColumnRenamed("doctor_name", "consultant_name").show()

# 15. Show patients from Hyderabad
patients.filter(col("city") == "Hyderabad").show()

# 16. Show female patients
patients.filter(col("gender") == "Female").show()

# 17. Show patients older than 35
patients.filter(col("age") > 35).show()

# 18. Show doctors from Hyderabad
doctors.filter(col("city") == "Hyderabad").show()

# 19. Show Cardiology doctors
doctors.filter(col("specialization") == "Cardiology").show()

# 20. Show doctors whose fee is greater than 1000
doctors.filter(col("consultation_fee") > 1000).show()
```

```python
# 21. Sort patients by age ascending
patients.orderBy("age").show()

# 22. Sort patients by age descending
patients.orderBy(col("age").desc()).show()

# 23. Show top 5 oldest patients
patients.orderBy(col("age").desc()).show(5)

# 24. Show youngest 3 patients
patients.orderBy("age").show(3)

# 25. Sort doctors by consultation fee descending
doctors.orderBy(col("consultation_fee").desc()).show()

# 26. Show top 3 highest fee doctors
doctors.orderBy(col("consultation_fee").desc()).show(3)

# 27. Show lowest fee doctors
doctors.orderBy("consultation_fee").show()

# 28. Sort appointments by appointment date
appointments.orderBy("appointment_date").show()

# 29. Sort bills by bill amount descending
bills.orderBy(col("bill_amount").desc()).show()

# 30. Show top 5 highest bills
bills.orderBy(col("bill_amount").desc()).show(5)
```

```python
from pyspark.sql.functions import count, avg, max, min, sum

# 31. Count patients by city
patients.groupBy("city").count().show()

# 32. Count patients by gender
patients.groupBy("gender").count().show()

# 33. Count doctors by specialization
doctors.groupBy("specialization").count().show()

# 34. Find average patient age
patients.select(avg("age")).show()

# 35. Find maximum patient age
patients.select(max("age")).show()

# 36. Find minimum patient age
patients.select(min("age")).show()

# 37. Find average consultation fee
doctors.select(avg("consultation_fee")).show()

# 38. Find maximum consultation fee
doctors.select(max("consultation_fee")).show()

# 39. Find total bill amount
bills.select(sum("bill_amount")).show()

# 40. Find average bill amount
bills.select(avg("bill_amount")).show()
```

```python
from pyspark.sql.functions import lit, concat, when

# 61. Add age_group column using age
patients.withColumn("age_group",
    when(col("age") < 30, "Young")
    .when(col("age") < 40, "Adult")
    .otherwise("Senior")
).show()

# 62. Add hospital_name
patients.withColumn("hospital_name", lit("BotCampus Hospital")).show()

# 63. Add fee_with_tax
doctors.withColumn("fee_with_tax", col("consultation_fee") * 1.18).show()

# 64. Add bill_with_tax
bills.withColumn("bill_with_tax", col("bill_amount") * 1.18).show()

# 65. Add bill_in_thousands
bills.withColumn("bill_in_thousands", col("bill_amount") / 1000).show()

# 66. Add country
patients.withColumn("country", lit("India")).show()

# 67. Add doctor_label
doctors.withColumn("doctor_label",
    concat(col("doctor_name"), lit(" - "), col("specialization"))
).show()

# 68. Add patient_label
patients.withColumn("patient_label",
    concat(col("name"), lit(" - "), col("city"))
).show()

# 69. Add high_bill_flag
bills.withColumn("high_bill_flag", col("bill_amount") > 1500).show()

# 70. Add senior_patient_flag
patients.withColumn("senior_patient_flag", col("age") > 40).show()
```

```python
from pyspark.sql.functions import when, col

# 71. Create patient age category: Young, Adult, Senior
patients.withColumn("age_category",
    when(col("age") < 30, "Young")
    .when(col("age") < 50, "Adult")
    .otherwise("Senior")
).show()

# 72. Create bill category: High, Medium, Low
bills.withColumn("bill_category",
    when(col("bill_amount") >= 1500, "High")
    .when(col("bill_amount") >= 1000, "Medium")
    .otherwise("Low")
).show()

# 73. Create doctor fee category: Premium, Standard, Basic
doctors.withColumn("fee_category",
    when(col("consultation_fee") >= 1500, "Premium")
    .when(col("consultation_fee") >= 1000, "Standard")
    .otherwise("Basic")
).show()

# 74. Create appointment priority
appointments.withColumn("priority",
    when(col("status") == "Pending", "High")
    .when(col("status") == "Completed", "Normal")
    .otherwise("Low")
).show()

# 75. Create city zone
patients.withColumn("zone",
    when(col("city").isin("Hyderabad","Bangalore","Chennai"), "South")
    .otherwise("Metro")
).show()
```

```python
from pyspark.sql.functions import to_date, year, month, datediff, current_date

# 76. Convert registration_date to date type
patients = patients.withColumn("registration_date", to_date("registration_date"))

# 77. Extract registration year
patients.withColumn("year", year("registration_date")).show()

# 78. Extract registration month
patients.withColumn("month", month("registration_date")).show()

# 79. Convert appointment_date to date type
appointments = appointments.withColumn("appointment_date", to_date("appointment_date"))

# 80. Extract appointment month
appointments.withColumn("month", month("appointment_date")).show()

# 81. Count appointments by appointment date
appointments.groupBy("appointment_date").count().show()

# 82. Count appointments by appointment month
appointments.groupBy(month("appointment_date")).count().show()

# 83. Find patients registered after 2023-06-01
patients.filter(col("registration_date") > "2023-06-01").show()

# 84. Find days since patient registration
patients.withColumn("days_since_reg",
    datediff(current_date(), col("registration_date"))
).show()

# 85. Find days since appointment date
appointments.withColumn("days_since_app",
    datediff(current_date(), col("appointment_date"))
).show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, row_number, dense_rank, sum as _sum, col, lit

# 86. Rank patients by age within city
w = Window.partitionBy("city").orderBy(col("age").desc())
patients.withColumn("rank", rank().over(w)).show()

# 87. Oldest patient per city
patients.withColumn("row_num", row_number().over(w)).filter(col("row_num") == 1).show()

# 88. Rank doctors by fee within specialization
w2 = Window.partitionBy("specialization").orderBy(col("consultation_fee").desc())
doctors.withColumn("dense_rank", dense_rank().over(w2)).show()

# 89. Highest fee doctor per specialization
doctors.withColumn("rank", rank().over(w2)).filter(col("rank") == 1).show()

# 90. Top 2 oldest patients per city
patients.withColumn("rank", row_number().over(w)).filter(col("rank") <= 2).show()

# 91. Rank doctors by consultation fee overall
doctors.withColumn("rank",
    rank().over(Window.orderBy(col("consultation_fee").desc()))
).show()

# 92. Rank cities by patient count
patients.groupBy("city").count() \
    .withColumn("rank", rank().over(Window.orderBy(col("count").desc()))).show()

# 93. Rank doctors by appointment count
appointments.groupBy("doctor_id").count() \
    .withColumn("rank", rank().over(Window.orderBy(col("count").desc()))).show()

# 94. Running total bill amount by payment mode
bills.withColumn("running_total",
    _sum("bill_amount").over(Window.partitionBy("payment_mode").orderBy("bill_amount"))
).show()

# 95. Running appointment count by doctor
appointments.withColumn("running_count",
    _sum(lit(1)).over(Window.partitionBy("doctor_id").orderBy("appointment_date"))
).show()
```

```python
# Create RDD
logs = spark.sparkContext.textFile("hospital_logs.txt")

# 96. Count total log records
print("Total logs:", logs.count())

# 97. Extract user names
users = logs.map(lambda x: x.split()[0])
print(users.collect())

# 98. Extract actions
actions = logs.map(lambda x: x.split()[1])
print(actions.collect())

# 99. Find unique users
print(users.distinct().collect())

# 100. Count each action
print(actions.map(lambda x: (x,1)).reduceByKey(lambda a,b: a+b).collect())
```

```python
# Import required functions
from pyspark.sql.functions import col, explode, array_contains

# 1. Read the JSON file
profiles = spark.read.json("patient_profiles.json", multiLine=True)

# (Make sure patients DataFrame is already loaded)
# patients = spark.read.csv("patients.csv", header=True, inferSchema=True)

# 2. Print schema
profiles.printSchema()

# 3. Extract patient_id, name, email, and phone
profiles.select(
    col("patient_id"),
    col("name"),
    col("contact.email").alias("email"),
    col("contact.phone").alias("phone")
).show()

# 4. Explode allergies
profiles.select(
    col("patient_id"),
    col("name"),
    explode(col("allergies")).alias("allergy")
).show()

# 5. Count patients by allergy
profiles.select(explode(col("allergies")).alias("allergy")) \
        .groupBy("allergy") \
        .count() \
        .show()

# 6. Find patients with Dust allergy
profiles.filter(array_contains(col("allergies"), "Dust")) \
        .select("patient_id", "name") \
        .show()

# 7. Find unique allergies
profiles.select(explode(col("allergies")).alias("allergy")) \
        .distinct() \
        .show()

# 8. Join profiles with patients data
joined_df = profiles.join(patients, "patient_id")

# 9. Show patient name, city, email, and allergy
p1 = profiles.alias("p1")
p2 = patients.alias("p2")

p1.join(p2, "patient_id") \
  .select(
      col("p2.name").alias("patient_name"),
      col("p2.city"),
      col("p1.contact.email").alias("email"),
      explode(col("p1.allergies")).alias("allergy")
  ) \
  .show()

# 10. Count allergies per city
p1.join(p2, "patient_id") \
  .select(
      col("p2.city"),
      explode(col("p1.allergies")).alias("allergy")
  ) \
  .groupBy("city", "allergy") \
  .count() \
  .show()
```
